```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sys,os
```

```python
basepath =  "/global/projecta/projectdirs/lsst/groups/PZ/protoDC2"
infile = "5pct_protoDC2catalog.dat"
#infile = "5pct_protoDC2_catalog_with_photom_errors.dat"
fullpath = os.path.join(basepath,infile)
data = np.loadtxt(fullpath,skiprows=1)
```

#truth catalog: protoDC2catalog.dat
#galaxy_id ra dec truez u uerr g gerr r rerr i ierr z zerr y yerr

```python
id = data[:,0]
sz = data[:,3]
u = data[:,4]
#uerr = data[:,5]
g = data[:,5]
#gerr = data[:,7]
r = data[:,6]
#rerr = data[:,9]
i = data[:,7]
#ierr = data[:,11]
z = data[:,8]
#zerr = data[:,13]
y = data[:,9]
#yerr = data[:,15]
```

```python
ug = u - g
gr = g - r
ri = r-i
iz = i-z
zy = z-y
```

```python
fig = plt.figure(figsize=(14,14))
plt.scatter(sz,i,marker='.',s=1,c='b')
plt.xlabel('redshift',fontsize=22)
plt.ylabel('imag',fontsize=22)
plt.ylim((15,30))
plt.plot([0.,1.],[25.3,25.3],color='r',label = "gold sample")
plt.plot([0.,1.],[26.8,26.8],color='r',linestyle='--',label = "i<26.8")
plt.legend(loc="upper left",fontsize=17)
```

```python
brightmask = (i<22.)
brightgr = gr[brightmask]
brightri = ri[brightmask]
brightsz = sz[brightmask]
goldmask = (i<=25.3)
goldgr = gr[goldmask]
goldri = ri[goldmask]
goldsz = sz[goldmask]
```

```python
plt.figure(figsize=(14,14))
plt.scatter(goldgr,goldri,marker='.',s=1,c='b',label = "gold sample")
plt.scatter(brightgr,brightri,marker='.',s=1,c='r',label="i<22")
plt.xlabel('g-r',fontsize=22)
plt.ylabel('r-i',fontsize=22)
plt.xlim((-1,3))
plt.ylim((-.5,1.75))
plt.legend(loc="upper left",fontsize=15)
```

```python
plt.figure(figsize=(14,14))
plt.scatter(goldsz,goldgr,marker='.',s=1,c='b',label = "gold sample g-r")
plt.scatter(brightsz,brightgr,marker='.',s=1,c='r',label="i<22 g-r")
plt.xlabel('redshift',fontsize=22)
plt.ylabel('g-r',fontsize=22)
plt.ylim((-.8,2.25))
plt.legend(loc="upper left",fontsize=15)
```

```python
plt.figure(figsize=(14,14))
plt.scatter(goldsz,goldri,marker='.',s=1,c='b',label = "gold sample r-i")
plt.scatter(brightsz,brightri,marker='.',s=1,c='r',label="i<22 r-i")
plt.xlabel('redshift',fontsize=22)
plt.ylabel('r-i',fontsize=22)
plt.ylim((-.5,1.3))
plt.legend(loc="upper left",fontsize=15)
```
